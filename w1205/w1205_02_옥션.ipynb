# 1~5페이지 까지 웹스크래핑을 해서 후기:50 이상, 구매 100건 이상, 가격 500,000원 이상
# CSV 파일로 저장( 이미지 링크, 제목, 가격, 후기, 구매 )

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
from bs4 import BeautifulSoup
import time
import os
import random
import csv
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)
options.add_argument("--disable-blink-features=AutomationControlled")
```

```python

for i in range(1,6):
    
    url = f"https://www.gmarket.co.kr/n/search?spm=gmktpc.home.searchtop.dsearchbox.4de1486axM4di2&keyword=tv&p={i}"
    
    browser = webdriver.Chrome(options=options) # 창 열기
    browser.maximize_window() # 창 최대로 확대

    # 페이지 열기
    browser.get(url)
    
    soup = BeautifulSoup(browser.page_source,"lxml")

    divs_main = soup.find("div",{"id":"section__inner-content-body-container"})
    divs = divs_main.find_all("div",{"class":"box__item-container"})

    f = open("gmarket.csv","a",encoding="utf-8-sig",newline="")
    writer = csv.writer(f)

    for div in divs:
        g1 = div.find("div",{"class":"box__image"}).img['src']
        g2 = div.find("div",{"class":"box__item-title"}).find("span",{"class":"text__item"}).text.strip()
        g3 = int(div.find("div",{"class":"box__price-seller"}).find("strong",{"class":"text text__value"}).text.strip().replace(",",""))
        if div.find("li",{"class":"list-item list-item__feedback-count"}):
            g4 = int(div.find("li",{"class":"list-item list-item__feedback-count"}).find("span",{"class":"text"}).text.strip().replace(",","")[1:-1])
        else:
            g4 = 0
            
        if div.find("li",{"class":"list-item list-item__pay-count"}):
            
            if div.find("li",{"class":"list-item list-item__pay-count"}).find("span",{"class":"text"}).text.strip().replace(",","").replace("구매","").isdigit():
                g5 = int(div.find("li",{"class":"list-item list-item__pay-count"}).find("span",{"class":"text"}).text.strip().replace(",","").replace("구매",""))
        else:
            g5 = 0
        
        # 1~5페이지 까지 웹스크래핑을 해서 후기:50 이상, 구매 100건 이상, 가격 500,000원 이상
        
        writer.writerow([g1,g2,g3,g4,g5])
        
        # if g4 >= 50 and g5 >= 100 and g3 >= 500000:
        #     writer.writerow([g1,g2,g3,g4,g5])

    f.close()
    
    browser.close()
```
