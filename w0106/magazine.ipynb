```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
from bs4 import BeautifulSoup
import time
import os
import random
import csv
import re
```

```python
url = 'https://www.bluer.co.kr/magazine'
browser = webdriver.Chrome()
browser.maximize_window()
browser.get(url)
```

```python
# 스크롤 내리기
pre_height = browser.execute_script("return document.body.scrollHeight")

while True:
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    
    if browser.find_element(By.XPATH,'//*[@id="btn-more"]').is_displayed():
        browser.find_element(By.XPATH,'//*[@id="btn-more"]').click()
        time.sleep(0.5)
    
    curr_height = browser.execute_script("return document.body.scrollHeight")
    if curr_height == pre_height:
        print("스크롤 완료")
        break
    
    pre_height = curr_height
    print(pre_height)
```

```python
soup = BeautifulSoup(browser.page_source,"lxml")

with open("magazine.html","w",encoding="utf8") as f:
    f.write(soup.prettify())
```

```python
with open("magazine.html","r",encoding="utf8") as f:
    soup = BeautifulSoup(f,"lxml")

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36"}

ul = soup.find("ul",{"class":"list-report"})
lis = ul.find_all("li",{"class":"magazine-thumb-item"})

f = open("magazine.csv","w",encoding="utf-8-sig",newline="")
writer = csv.writer(f)

for li in lis:
    id = li.find("a",{"class":"br-card br-card-1"})['href'].replace('/magazine/','')
    print(id)
    img = "".join(re.findall(r'\((.*?)\)',li.find("div",{"class":"embed-responsive-item bg-cover"})['style']))
    imga = f'https://www.bluer.co.kr{img}'
    print(imga)
    category = li.find("span",{"class":"card-category"}).text.strip()
    print(category)
    title = li.find("h4",{"class":"card-title"}).text.strip()
    print(title)
    description = li.find("p",{"class":"card-description"}).text.strip().replace("\n","")
    print(description)

    writer.writerow([id,f"magazine/{id}.jpg",category,title,description])

    
    # img_req = requests.get(imga,headers=headers)
    # os.makedirs("magazin",exist_ok=True) # 폴더생성    
    # with open(f"./magazin/{id}.jpg","wb") as f:
    #     f.write(img_req.content) # 파일로 저장
    
f.close()
```

```python

f = open("./magazine3.csv","r",encoding="utf-8-sig")
rdr = csv.reader(f)

f2 = open("magazine4.csv","w",encoding="utf-8-sig",newline="")
writer = csv.writer(f2)

for line in rdr:
    
    url = f'https://www.bluer.co.kr/magazine/{line[0]}'
    browser = webdriver.Chrome()
    browser.maximize_window()
    browser.get(url)
    soup = BeautifulSoup(browser.page_source,"lxml")

    context = soup.find("div",{"class":"document"})
    
    writer.writerow([line[0],context])

    browser.close()
    
f.close()
f2.close()
```
