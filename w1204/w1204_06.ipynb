```python
# 화면이 로링될때까지 대기
# time.sleep(7)

WebDriverWait(browser,10).until(EC.presence_of_all_elements_located(By.XPATH,'//*[@id="__next"]/div/main/div[4]/div/div[2]/div[2]/div[1]'))
```

```python
# 화면스크롤 내림
pre_height = browser.execute_script("return document.body.scrollHeight")
pre_height
```

```python
while True:
    # 스크롤 아래로 내림
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    
    # 추가 데이터를 불러오는동안 대기
    time.sleep(3)
    curr_height = browser.execute_script("return document.body.scrollHeight")
    if curr_height == pre_height:
        break
    pre_height = curr_height
    print("높이 변화 : ",pre_height)

print("스크롤 완료 : ",curr_height)
    
```

```python
soup = BeautifulSoup(browser.page_source,"lxml")

domResult = soup.find("div",{"class":"domestic_results__gp5WB"})
fl = domResult.find_all("div",{"class":"domestic_Flight__8bR_b"})
```

```python
for i,f in enumerate(fl):
    
    print(f.find("b",{"class":"airline_name__0Tw5w"}).text.strip())
    print(f.find_all("b",{"class":"route_time__xWu7a"})[0].text.strip())
    print(f.find_all("b",{"class":"route_time__xWu7a"})[1].text.strip())
    print(f.find("i",{"class":"domestic_num__ShOub"}).text.strip())
    print("-"*80)
```

```python
# CSV파일로 저장
import csv

title = ['항공사명','출발시간','도착시간','금액']

fo = open("airline.csv","w",encoding="utf-8-sig",newline="")
writer = csv.writer(fo)
writer.writerow(title)

for i,f in enumerate(fl):
    
    n1 = f.find("b",{"class":"airline_name__0Tw5w"}).text.strip()
    n2 = f.find_all("b",{"class":"route_time__xWu7a"})[0].text.strip()
    n3 = f.find_all("b",{"class":"route_time__xWu7a"})[1].text.strip()
    n4 = int(f.find("i",{"class":"domestic_num__ShOub"}).text.strip().replace(",",""))
    
    # wr = [n1,n2,n3,n4]

    writer.writerow([n1,n2,n3,n4])
   
fo.close()    
```
